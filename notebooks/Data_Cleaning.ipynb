Handling Missing Genus, Order, and Family Values

```python
def taxon_majority_map(df, taxon_col, target_col):
    """
    Returns a mapping from Taxon -> most frequent non-null value of target_col.
    """
    return (
        df.dropna(subset=[target_col])
          .groupby(taxon_col)[target_col]
          .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
    )
```

```python
order_map  = taxon_majority_map(df, 'Taxon', 'Order')
family_map = taxon_majority_map(df, 'Taxon', 'Family')
genus_map  = taxon_majority_map(df, 'Taxon', 'Genus')
```

```python
df['Order'] = df['Order'].fillna(df['Taxon'].map(order_map))
df['Family'] = df['Family'].fillna(df['Taxon'].map(family_map))
df['Genus'] = df['Genus'].fillna(df['Taxon'].map(genus_map))
```

```python
df[['Order', 'Family', 'Genus']] = (
    df[['Order', 'Family', 'Genus']]
    .fillna('Unknown')
)
```

```python
df['order_imputed']  = df['Order'].eq('Unknown').astype(int)
df['family_imputed'] = df['Family'].eq('Unknown').astype(int)
df['genus_imputed']  = df['Genus'].eq('Unknown').astype(int)
```

Feature Encoding

Map Appendix Categories to Numeric Codes

```python
appendix_map = {'I': 3, 'II': 2, 'III': 1}
df['App.'] = df['App.'].map(appendix_map)
```

Strip Leading and Trailing Whitespace in Columns

```python
for col in ['Taxon','Order','Family','Genus']:
    df[col] = df[col].str.strip().str.title()
```

```python
for col in ['Importer','Exporter','Origin']:
    df[col] = df[col].str.strip()
```

```python
df['Term'] = df['Term'].str.upper().str.strip()
```

Clip Importer and Exporter Quantities to Non-Negative Values

```python
df['Importer reported quantity'] = df['Importer reported quantity'].clip(lower=0)
df['Exporter reported quantity'] = df['Exporter reported quantity'].clip(lower=0)
```

Create is_live Indicator Column

```python
df['Term'] = df['Term'].str.upper().str.strip()

df['is_live'] = (df['Term'] == 'LIVE').astype(int)
```

Log Transform Quantity Columns

```python
df['import_qty_log'] = np.log1p(df['Importer reported quantity'])
df['export_qty_log'] = np.log1p(df['Exporter reported quantity'])
```

Purpose and Source Columns Encoded

```python
purpose_risk_map = {
    'T': 3,  # Commercial
    'H': 3,  # Hunting trophy
    'P': 2,  # Personal
    'Q': 2,  # Circus / exhibitions
    'M': 2,  # Medical
    'B': 1,  # Breeding
    'Z': 1,  # Zoo
    'G': 1,  # Botanical garden
    'S': 1,  # Scientific
    'E': 1,  # Educational
    'N': 1,  # Reintroduction
    'L': 1,   # Law enforcement
    'Unknown': 2
}

df['Purpose'] = df['Purpose'].map(purpose_risk_map).fillna(1)
```

```python
source_risk_map = {
    'W': 3,  # Wild
    'I': 3,  # Confiscations / seizures
    'U': 2,  # Source unknown
    'X': 2,  # Marine (jurisdiction unclear)
    'O': 1,  # Pre-convention
    'R': 1,  # Ranched
    'C': 1,  # Captive-bred
    'F': 1,  # Born in captivity
    'A': 1,  # Artificially propagated
    'D': 1,  # Captive/propagated (App I)
    'Y': 1,   # Assisted production
    'Unknown': 2
}

df['Source'] = df['Source'].map(source_risk_map).fillna(2)
```

Extract Top 6 Most Frequent Species

```python
df=pd.read_csv("Wildlife_Dataset_Cleaned.csv")
```

```python
combo_counts = (
    df.groupby(["Taxon", "Order", "Family", "Genus"])
      .size()
      .reset_index(name="count")
)

total_unique_combos = combo_counts.shape[0]

combo_counts, total_unique_combos
```

```python
top_6_combos = (
    df.groupby(["Taxon", "Order", "Family", "Genus"])
      .size()
      .reset_index(name="count")
      .sort_values("count", ascending=False)
      .head(6)
)

top_6_combos
```

```python
df_top_6_rows = df.merge(
    top_6_combos[["Taxon", "Order", "Family", "Genus"]],
    on=["Taxon", "Order", "Family", "Genus"],
    how="inner"
)
```

```python
df_top_6_rows.to_csv("Wildlife_Dataset_Top6.csv", index=False)
from google.colab import files
files.download("Wildlife_Dataset_Top6.csv")
```
